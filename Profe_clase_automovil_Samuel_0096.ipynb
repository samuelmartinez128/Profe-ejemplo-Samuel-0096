```python
# clase Automovil, atributos, marca, placa, precio
# funciones leerdatos(), mostrardatos()
# Crear 3 objetos carro1, carro2. carro3
class Automovil:
  def __init__(self, marca, placa, precio):
    self.marca = marca
    self.placa = placa
    self.precio = precio

  def leerdatos(self):
    self.marca = input("Ingrese la marca del automovil: ")
    self.placa = input("Ingrese la placa del automovil: ")
    self.precio = input("Ingrese el precio del automovil: ")
    return self.marca, self.placa, self.precio

  def mostrardatos(self):
    print("Marca: ", self.marca)
    print("Placa: ", self.placa)
    print("Precio: ", self.precio)
    return self.marca, self.placa, self.precio
    # Fin de la clase automovil

#Creando los objetos carro1, carro2, carro3
print("carro numero 1")
carro1 = Automovil("", "", "") # Creando objeto carro1
carro1.leerdatos()
carro1.mostrardatos()
print("-------------------")
print("carro numero 2")
carro2 = Automovil("", "", "") # Creando objeto carro2
carro2.leerdatos()
carro2.mostrardatos()
print("-------------------")
print("carro numero 3")
carro3 = Automovil("", "", "") # Creando onjeto carrro3
carro3.leerdatos()
carro3.mostrardatos()
print("Software desrrollado por Samuel Martinez NL = 35")
```
